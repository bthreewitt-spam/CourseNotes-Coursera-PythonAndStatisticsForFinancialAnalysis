```python
import pandas as pd
```

```python
die = pd.DataFrame([1,2,3,4,5,6])
trial =  50
results = [die.sample(2, replace=True).sum().loc[0] for i in range(trial)]
results[:10]
```

```python
freq = pd.DataFrame(results) [0].value_counts() # count values
sort_freq = freq.sort_index() # sort index
sort_freq
```

```python
sort_freq.plot(kind='bar', color='blue')
```

## Relative Frequency = Frequency / Num of Trials

```python
relative_freq = sort_freq/trial
relative_freq.plot(kind='bar', color='blue')
```

```python

def roll_dice(num_trials):
    result = [die.sample(2, replace=True).sum().loc[0] for i in range(num_trials)]
    return pd.DataFrame(result) [0].value_counts().sort_index()
```

```python
roll_dice(10_000).plot(kind='bar', color='blue')
```
