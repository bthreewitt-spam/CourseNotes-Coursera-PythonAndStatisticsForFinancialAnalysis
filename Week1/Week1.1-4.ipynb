```python
meta['Next Close'] = meta['Close'].shift(-1)    #Create new column "Next Close" and populate it
meta[['Close', 'Next Close']]    #View Results
```

```python
meta['PriceDiff'] = meta['Next Close'] - meta['Close']
meta[['Close', 'PriceDiff']]
```

```python
meta['Return'] = meta['PriceDiff']/meta['Close']
meta[['Close', 'Return']]
```

```python
meta['Direction'] = [1 if meta.loc[date, 'PriceDiff'] > 0 else -1 for date in meta.index]
meta[['Close', 'Direction']]
```

```python
# meta['MA3'] = (meta['Close'] + meta['Close'].shift(1) + meta['Close'].shift(2)) / 3
# better way todo it

meta['MA3'] = meta['Close'].rolling(3).mean()
meta[['Close', 'MA3']]
```

# .shift
- 1 = column down
- -1 = column up

```python
meta['MA40'] = meta['Close'].rolling(40).mean()
meta['MA200'] = meta['Close'].rolling(200).mean()
meta[['Close', 'MA40', 'MA200']]
```

```python
# meta['Close'].plot()
# meta['MA40'].plot()
# meta['MA200'].plot()

# better way todo it
meta[['Close', 'MA40', 'MA200']].plot()
```

- MA40 = Fast Signal
- MA200 = Slow Signal

ma40 > ma200 = buy signal
